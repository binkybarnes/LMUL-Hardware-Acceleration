```python
import matplotlib.pyplot as plt
import numpy as np

#yoinked data from the LSTM_test notebook; might make that write into a .txt file later so that this can be automated.
datasets = ["FashionMNIST", "KMNIST", "SeqMNIST"]

fp32 = {
    "FashionMNIST": [84.610, 84.340, 84.200, 83.600, 84.730],
    "KMNIST":       [85.640, 85.530, 85.080, 86.480, 86.600],
    "SeqMNIST":     [96.610, 96.470, 96.090, 95.680, 96.960],
}

lmul = {
    "FashionMNIST": [83.550, 83.060, 83.910, 82.820, 84.010],
    "KMNIST":       [85.320, 85.460, 85.290, 86.490, 86.600],
    "SeqMNIST":     [96.400, 96.550, 96.180, 95.470, 96.920],
}

#avgs
fp32_avg = [sum(fp32[d]) / len(fp32[d]) for d in datasets]
lmul_avg = [sum(lmul[d]) / len(lmul[d]) for d in datasets]
difference_avg = [a - b for (a, b) in zip(fp32_avg, lmul_avg)]
x = np.arange(len(datasets))
width = 0.35

plt.figure(figsize=(8,5))
#plt.bar(x - width/2, fp32_avg, width, label="FP32 Avg")
#plt.bar(x + width/2, lmul_avg, width, label="LMUL Avg")
plt.bar(x, difference_avg, width, label="Difference between FP32 and LMUL avg Score")
plt.xticks(x, datasets)
plt.ylabel("Accuracy Difference (%)")
plt.title("Average Accuracy Difference Across 5 Seeds (LSTM Classifier)")
#plt.legend()
plt.grid(axis="y")
plt.ylim(0, 2)
plt.show()
```
